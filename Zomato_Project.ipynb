
#**zomato data analysis**

step-1 Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

step-2

```python
dataframe=pd.read_csv("/content/Zomato data .csv")
```

```python
print(dataframe)
```

convert the datatype of column - rate


```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)


dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

#Type of restrurant


```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.ylabel("Type of Restaurant")  # Correct spelling of xlabel
plt.show()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("type of restaurant", c="red", size=20)
plt.ylabel("votes", c="red",size=20)
plt.show()
```

concusion - dinning restrurent has recieved maximum votes


```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=20)
plt.title("ratings distribution")
plt.show()
```

Conclusion - the mejority resturant received ratings from 3.5 to 4

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# which mode recives maximum rating


```python
dataframe.head()
```

```python
plt.figure(figsize= (10,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

Conclusion - offline order recived lower rating in comparision to online order

```python
dataframe.head()
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("heatmap")
plt.xlabel('online order')
plt.ylabel('listed in (type)')
plt.show()
```

Conclusion - dining restaurent primarily accept offline orders, where as cafes primarily recived online orders , this suggest that client preffence orders in person at restaurants, but prefer online ordering at cafes
